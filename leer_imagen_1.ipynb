Importamos las librerias necesarias para trabajar con las imagenes

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

Traemos la imagen, cargamos en un archivo y la mostramos con plt

```python
imagen = cv2.imread('/content/drive/MyDrive/Imagenes_ML/datos/imagen_1.png')
plt.imshow(imagen)
plt.show()
```

Mostramos el tipo de imagen ( array) y la dimensionalidad del mismo

```python
print('El tipo de imagen es : ', type(imagen))
print('Dimensionalidad de la imagen', imagen.shape)
```

Hacemos una copia de la imagen

```python
imagen_copy = np.copy(imagen)
```

cambiamos el color de la imagen

```python
imagen_copy = cv2.cvtColor(imagen_copy,cv2.COLOR_BGR2RGB)
print(imagen_copy.shape)
```

```python
plt.imshow(imagen_copy)
plt.show()
```

```python
imagen_copy = cv2.cvtColor(imagen_copy,cv2.COLOR_BGR2GRAY)
print(imagen_copy.shape)
```

```python
plt.imshow(imagen_copy)
plt.show()
```

```python
plt.imshow(imagen_copy,cmap ='gray')
print (imagen_copy.shape)
plt.show()
```
