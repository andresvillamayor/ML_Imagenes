```python
# Lectura de imagen
img2 = imread(imagen)
X = img2[:,:,0] # Sacamos todos los canales  RGB
plt.imshow(X)
plt.show ()
```

```python
# como saber los valores de la imagen
# definimos una funcion
def valorimagen(img2):
  print('Tamao de la imagen= ', img2.shape )
  print('Maximo =' ,np.max(img2))
  print('Minimo= ', np.min(img2))
```

```python
# para dejar solo en la escala de gris
img2 = imread(imagen)
X = img2[:,:,0] # Sacamos todos los canales  RGB
plt.imshow(X,cmap ='gray')
plt.show()
```

```python
 # valor de la imagen  desde la funcion valorimagen
 valorimagen(X)
```

```python
# Tenemos una imagen de 512 x 512 con un minimo de 0 a 255 en escala de grises
# Hacemos un Muestreo por tono de grises
p = 32  # Cantidad de pixel por tono de grises
(Nx,Mx) = X.shape  # creamos una matriz a partir del shape
Ny = Nx
My = Mx
Y  = np.zeros((Ny,My),np.uint8)
for i in range (Ny):
  for j in range(My):
    x = int(np.fix(X[i,j]/p)*p)
    Y [i,j] = x
cv2_imshow(Y)
valorimagen(Y)
print('La imagen tiene:', int(256/p),'tonos de grises')
```

```python
# Generamos un histograma con los valores anteriores de la escala de grises
# Generamos una funcion que nos devuelva la cantidad de gris
def imhist(X,n = 256):
  (N,M) = X.shape
  h = np.zeros((n,))
  for i in range(N):
    for j in range(M):
      x = X[i,j]
      h[x] = h[x] + 1
  return h
```

```python
# Dibujamos el histograma conlos valores
n = 256
h = imhist(Y,n=n)
plt.figure(figsize = (12,8))
plt.plot(range(n),h[0:n])
plt.show()
# si observamos el histograma tenemos 256/p en las cantidad de barras de grises
```
