Un histograma es una gráfica en donde se muestra la frecuencia con las que aparecen los distintos niveles de intensidad de una imagen a escala de grises, normalmente el nivel de intensidad está en el rango de 0 a 255, en donde el valor 0 representa los color negro y 255 el color blanco, utilizando el histograma de una imagen podemos modificar sus características, por ejemplo, el brillo y contraste, existen además otras aplicaciones que veremos más adelante.

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

Importamos una foto y lo colocamos con una escala de grises

```python
imagen = cv2.imread('/content/drive/MyDrive/Imagenes_ML/datos/imagen_2.png')
plt.imshow(imagen)
plt.show()
```

```python
imagen_copy = np.copy(imagen)
print ('Dimension de la imagen',imagen_copy.shape)
```

```python
hist = cv2.calcHist([imagen_copy], [0], None, [256], [0, 256])
plt.plot(hist, color='orange' )
plt.xlabel('intensidad de iluminacion')
plt.ylabel('cantidad de pixeles')
plt.show()
```

```python
imagen_copy2 = cv2.cvtColor(imagen_copy,cv2.COLOR_BGR2RGB)
print(imagen_copy2.shape)
plt.imshow(imagen_copy2)
plt.show()
```

```python
hist = cv2.calcHist([imagen_copy2], [0], None, [256], [0, 256])
plt.plot(hist, color='orange' )
plt.xlabel('intensidad de iluminacion')
plt.ylabel('cantidad de pixeles')
plt.show()
```
