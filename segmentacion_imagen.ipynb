```python
# Librerias
from cv2 import imread
from google.colab.patches import cv2_imshow
import matplotlib.pyplot as plt
import numpy as np
import cv2
```

```python
# leer la imagen - cargar la imagen para ver
imagen = cv2.imread('/content/drive/MyDrive/Imagenes_ML/datos/onerice.png')
plt.imshow(imagen)
plt.show()

#imagen = imread('/content/drive/MyDrive/Imagenes_ML/datos/onerice.png')
#cv2_imshow(imagen)
```

```python
# funcion de como es la imagen - tamaano - maximo de la imagen y minimo de la imagen
def howis(imagen):
  print('size =' , imagen.shape) # tamao de la imagen
  print('max = ', np.max(imagen)) # maximo de la imgen
  print('min =',np.min(imagen)) # minimo de la imagen

howis(imagen)
```

```python
# Selecionar el primer canal de la imagen
x = imagen[:,:,0] # todas las filas y columnas - canal 0
howis(x)
plt.imshow(x,cmap = 'gray')
plt.show()
#cv2_imshow(x)
```

```python
# Tresholding - Segmentacion Simple
# funcion de segmentacion por umbral
# ir por pixel por pixel para ver si supera el umbral de 170
def segmenta(x,t):
  (N,M) = x.shape
  y = np.zeros((N,M)) # imagen de salida tiene el mismo tamano que la imagen de salida
  area = 0
  for i in range(N):
    for j in range(M):
      if x[i,j] > t:
        y[i,j] = 255
        area = area + 1
  print('Area = ', area)
  return y
```

```python
# Segmentacion
y = segmenta (x,170) # valore de la funcion , se pasa la matriz y el threshold es 170
howis (y)
plt.imshow(y)
plt.show()
```
