```python
# Librerias
import cv2
import numpy as np
from  cv2 import imread
from google.colab.patches import cv2_imshow
import matplotlib.pyplot as plt
import pandas as pd
```

## Implementar t√©cnicas de mejoramiento de imagenes
### Aplicar el Algoritmo de mejora de contraste Adaptativo de histograma limitado(CLAHE)

```python
# Unir con el Google Drive
from google.colab import drive
drive.mount('/content/drive')
```

```python
# Leer la imagen en escala de grises
imagen = cv2.imread('/content/drive/MyDrive/Imagenes_ML/Final/Panoramic radiography database/1.jpg',0)
```

```python
# Aplicar CLAHE
clahe = cv2.createCLAHE(clipLimit = 2.0,tileGridSize = (8,8))
clahe_img = clahe.apply(imagen)
```

### Mostrar las imagen Original y la Mejorada con CLAHE

```python
# Mostrar la imagen Original y la imagen mejorada con CLAHE
# Imagen Original
plt.figure(figsize=(10,5))
plt.subplot (1,2,1)
plt.imshow(imagen,cmap='gray')
plt.title ('Imagen Original')
plt.axis('off')
# Imagen mejorada con CLAHE
plt.subplot (1,2,2)
plt.imshow(clahe_img,cmap='gray')
plt.title ('Imagen con CLAHE')
plt.axis('off')
#
plt.tight_layout()
plt.show()
```
